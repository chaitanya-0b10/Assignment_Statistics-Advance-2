```python
import random

int_list = [random.randint(90, 130) for _ in range(100)]
```

```python
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

mean_value = calculate_mean(int_list)
```

```python
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    midpoint = n // 2
    
    if n % 2 == 0:
        return (sorted_numbers[midpoint - 1] + sorted_numbers[midpoint]) / 2
    else:
        return sorted_numbers[midpoint]

median_value = calculate_median(int_list)
```

```python
from collections import Counter

def calculate_mode(numbers):
    count = Counter(numbers)
    mode = count.most_common(1)[0][0]
    return mode

mode_value = calculate_mode(int_list)
```

```python
def calculate_weighted_mean(values, weights):
    weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

weights = [random.randint(1, 10) for _ in range(100)]  # Example weights
weighted_mean_value = calculate_weighted_mean(int_list, weights)
```

```python
import math

def calculate_geometric_mean(numbers):
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

geometric_mean_value = calculate_geometric_mean(int_list)
```

```python
def calculate_harmonic_mean(numbers):
    reciprocal_sum = sum(1 / x for x in numbers)
    return len(numbers) / reciprocal_sum

harmonic_mean_value = calculate_harmonic_mean(int_list)
```

```python
def calculate_midrange(numbers):
    return (max(numbers) + min(numbers)) / 2

midrange_value = calculate_midrange(int_list)
```

```python
def calculate_trimmed_mean(numbers, percentage):
    n = len(numbers)
    trimmed_amount = int(n * percentage / 100)
    trimmed_numbers = sorted(numbers)[trimmed_amount:-trimmed_amount]
    return sum(trimmed_numbers) / len(trimmed_numbers)

trimmed_mean_value = calculate_trimmed_mean(int_list, 5) 

print(trimmed_mean_value)
```

```python
def calculate_variance(numbers):
    mean = sum(numbers) / len(numbers)
    variance = sum((x - mean) ** 2 for x in numbers) / (len(numbers) - 1)
    return variance

def calculate_std_dev(variance):
    return variance ** 0.5

variance_value = calculate_variance(int_list2)
std_dev_value = calculate_std_dev(variance_value)
print("Variance:", variance_value)
print("Standard Deviation:", std_dev_value)
```

```python
def calculate_iqr(numbers):
    sorted_numbers = sorted(numbers)
    q1 = np.percentile(sorted_numbers, 25)
    q3 = np.percentile(sorted_numbers, 75)
    iqr = q3 - q1
    return iqr

iqr_value = calculate_iqr(int_list2)
print("Interquartile Range (IQR):", iqr_value)
```

```python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
       
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
    
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
    
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.2, 0.2]
rv = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
```

```python
import random

class DieSimulation:
    def __init__(self, num_rolls):
        self.num_rolls = num_rolls
        self.outcomes = []

    def roll_die(self):
        return random.randint(1, 6)

    def simulate_rolls(self):
        self.outcomes = [self.roll_die() for _ in range(self.num_rolls)]

    def expected_value(self):
        return sum(self.outcomes) / self.num_rolls

    def variance(self):
        mean = self.expected_value()
        return sum((x - mean) ** 2 for x in self.outcomes) / self.num_rolls

num_rolls = 10000
simulation = DieSimulation(num_rolls)
simulation.simulate_rolls()
print("Expected Value:", simulation.expected_value())
print("Variance:", simulation.variance())
```

```python
import numpy as np
from scipy.stats import binom, poisson

def generate_samples(distribution, params, size):
    
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, size)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

binomial_mean, binomial_variance = generate_samples('binomial', (10, 0.5), 1000)
print("Binomial Distribution - Mean:", binomial_mean, "Variance:", binomial_variance)

poisson_mean, poisson_variance = generate_samples('poisson', (3,), 1000)
print("Poisson Distribution - Mean:", poisson_mean, "Variance:", poisson_variance)
```

```python
import numpy as np

def generate_normal_samples(mean, std_dev, size):
    """
    Generate random samples from a normal distribution and calculate their mean, variance, and standard deviation.
    
    :param mean: Mean of the normal distribution.
    :param std_dev: Standard deviation of the normal distribution.
    :param size: Number of samples to generate.
    :return: A tuple containing the mean, variance, and standard deviation of the samples.
    """
    samples = np.random.normal(loc=mean, scale=std_dev, size=size)
    
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

mean = 0
std_dev = 1
size = 1000

sample_mean, sample_variance, sample_std_dev = generate_normal_samples(mean, std_dev, size)
print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
print("Sample Standard Deviation:", sample_std_dev)
```

```python
import numpy as np
from scipy.stats import norm

old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

old_conversion_rate = old_layout.mean()
new_conversion_rate = new_layout.mean()

pooled_conversion_rate = (old_layout.sum() + new_layout.sum()) / (len(old_layout) + len(new_layout))

standard_error = np.sqrt(pooled_conversion_rate * (1 - pooled_conversion_rate) * (1/len(old_layout) + 1/len(new_layout)))

z_score = (new_conversion_rate - old_conversion_rate) / standard_error

p_value = 1 - norm.cdf(z_score)

print(f"Old Conversion Rate: {old_conversion_rate:.4f}")
print(f"New Conversion Rate: {new_conversion_rate:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The new layout has a significantly higher conversion rate.")
else:
    print("There is no significant difference in conversion rates between the old and new layouts.")
```

```python
import numpy as np
from scipy.stats import t

def ab_test(layout_a_clicks, layout_b_clicks):
    # Convert lists to numpy arrays
    layout_a_clicks = np.array(layout_a_clicks)
    layout_b_clicks = np.array(layout_b_clicks)
    
    # Calculate means
    mean_a = np.mean(layout_a_clicks)
    mean_b = np.mean(layout_b_clicks)
    
    # Calculate standard deviations
    std_a = np.std(layout_a_clicks, ddof=1)
    std_b = np.std(layout_b_clicks, ddof=1)
    
    # Calculate sample sizes
    n_a = len(layout_a_clicks)
    n_b = len(layout_b_clicks)
    
    # Calculate standard error
    se = np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))
    
    # Calculate t-statistic
    t_stat = (mean_a - mean_b) / se
    
    # Calculate degrees of freedom
    df = ((std_a**2 / n_a) + (std_b**2 / n_b))**2 / (((std_a**2 / n_a)**2 / (n_a - 1)) + ((std_b**2 / n_b)**2 / (n_b - 1)))
    
    # Calculate p-value
    p_value = 2 * t.cdf(-abs(t_stat), df)
    
    return t_stat, df, p_value

# Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test
t_stat, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

print(f"T-Statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("There is a significant difference between the two layouts.")
else:
    print("There is no significant difference between the two layouts.")
```

```python
import numpy as np
from scipy.stats import t

def analyze_drug_effectiveness(existing_drug_levels, new_drug_levels):
    # Convert lists to numpy arrays
    existing_drug_levels = np.array(existing_drug_levels)
    new_drug_levels = np.array(new_drug_levels)
    
    # Calculate means
    mean_existing = np.mean(existing_drug_levels)
    mean_new = np.mean(new_drug_levels)
    
    # Calculate standard deviations
    std_existing = np.std(existing_drug_levels, ddof=1)
    std_new = np.std(new_drug_levels, ddof=1)
    
    # Calculate sample sizes
    n_existing = len(existing_drug_levels)
    n_new = len(new_drug_levels)
    
    # Calculate standard error
    se = np.sqrt((std_existing**2 / n_existing) + (std_new**2 / n_new))
    
    # Calculate t-statistic
    t_stat = (mean_existing - mean_new) / se
    
    # Calculate degrees of freedom
    df = ((std_existing**2 / n_existing) + (std_new**2 / n_new))**2 / (((std_existing**2 / n_existing)**2 / (n_existing - 1)) + ((std_new**2 / n_new)**2 / (n_new - 1)))
    
    # Calculate p-value
    p_value = 2 * t.cdf(-abs(t_stat), df)
    
    return t_stat, df, p_value

# Data
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform analysis
t_stat, df, p_value = analyze_drug_effectiveness(existing_drug_levels, new_drug_levels)

print(f"T-Statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The new drug is significantly more effective in reducing cholesterol levels.")
else:
    print("There is no significant difference in effectiveness between the new and existing drugs.")
```

```python
import numpy as np
from scipy.stats import t

def analyze_intervention(pre_scores, post_scores):
    # Convert lists to numpy arrays
    pre_scores = np.array(pre_scores)
    post_scores = np.array(post_scores)
    
    # Calculate means
    mean_pre = np.mean(pre_scores)
    mean_post = np.mean(post_scores)
    
    # Calculate standard deviations
    std_pre = np.std(pre_scores, ddof=1)
    std_post = np.std(post_scores, ddof=1)
    
    # Calculate sample sizes
    n_pre = len(pre_scores)
    n_post = len(post_scores)
    
    # Calculate standard error
    se = np.sqrt((std_pre**2 / n_pre) + (std_post**2 / n_post))
    
    # Calculate t-statistic
    t_stat = (mean_post - mean_pre) / se
    
    # Calculate degrees of freedom
    df = ((std_pre**2 / n_pre) + (std_post**2 / n_post))**2 / (((std_pre**2 / n_pre)**2 / (n_pre - 1)) + ((std_post**2 / n_post)**2 / (n_post - 1)))
    
    # Calculate p-value
    p_value = 2 * t.cdf(-abs(t_stat), df)
    
    return t_stat, df, p_value

# Data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform analysis
t_stat, df, p_value = analyze_intervention(pre_intervention_scores, post_intervention_scores)

print(f"T-Statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("The intervention had a significant impact on math scores.")
else:
    print("The intervention did not have a significant impact on math scores.")
```

```python
import numpy as np
from scipy.stats import t

def analyze_salary_gap(male_salaries, female_salaries):
    # Convert lists to numpy arrays
    male_salaries = np.array(male_salaries)
    female_salaries = np.array(female_salaries)
    
    # Calculate means
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)
    
    # Calculate standard deviations
    std_male = np.std(male_salaries, ddof=1)
    std_female = np.std(female_salaries, ddof=1)
    
    # Calculate sample sizes
    n_male = len(male_salaries)
    n_female = len(female_salaries)
    
    # Calculate standard error
    se = np.sqrt((std_male**2 / n_male) + (std_female**2 / n_female))
    
    # Calculate t-statistic
    t_stat = (mean_male - mean_female) / se
    
    # Calculate degrees of freedom
    df = ((std_male**2 / n_male) + (std_female**2 / n_female))**2 / (((std_male**2 / n_male)**2 / (n_male - 1)) + ((std_female**2 / n_female)**2 / (n_female - 1)))
    
    # Calculate p-value
    p_value = 2 * t.cdf(-abs(t_stat), df)
    
    return t_stat, df, p_value

# Generate synthetic salary data
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Perform analysis
t_stat, df, p_value = analyze_salary_gap(male_salaries, female_salaries)

print(f"T-Statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("There is a significant difference between the average salaries of male and female employees.")
else:
    print("There is no significant difference between the average salaries of male and female employees.")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate the data
np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])
print(contingency_table)

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-Value: {p:.4f}")

if p < 0.05:
    print("There is a significant association between age groups and voter preferences.")
else:
    print("There is no significant association between age groups and voter preferences.")
```
